BoxPlots


```python
 # Function to remove outliers using IQR
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# List of numerical variables for boxplots
numerical_vars = [
    'Age', 'MonthlyIncome', 'HourlyRate', 'DailyRate', 'MonthlyRate',
    'PercentSalaryHike', 'DistanceFromHome', 'YearsAtCompany',
    'YearsSinceLastPromotion', 'TotalWorkingYears', 'NumCompaniesWorked',
    'TrainingTimesLastYear', 'YearsWithCurrManager'
]

# Plot boxplots without outliers
fig, axes = plt.subplots(nrows=7, ncols=2, figsize=(16, 28))
axes = axes.flatten()

for i, var in enumerate(numerical_vars):
    cleaned_df = remove_outliers_iqr(df, var)
    sns.boxplot(y=cleaned_df[var], ax=axes[i], color='lightgreen')
    axes[i].set_title(f'Boxplot of {var} (Outliers Removed)')
    axes[i].set_xlabel('')
    axes[i].set_ylabel(var)

# Hide unused subplot if odd number of plots
if len(numerical_vars) % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()

```

## Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import precision_recall_curve
```

```python
def model(classifier,x_train,y_train,x_test,y_test):
    
    classifier.fit(x_train,y_train)
    prediction = classifier.predict(x_test)
    cv = RepeatedStratifiedKFold(n_splits = 10,n_repeats = 3,random_state = 1)
    print("Cross Validation Score : ",'{0:.2%}'.format(cross_val_score(classifier,x_train,y_train,cv = cv,scoring = 'roc_auc').mean()))
def model_evaluation(classifier,x_test,y_test):
    
    # Confusion Matrix
    cm = confusion_matrix(y_test,classifier.predict(x_test))
    names = ['True Neg','False Pos','False Neg','True Pos']
    counts = [value for value in cm.flatten()]
    percentages = ['{0:.2%}'.format(value) for value in cm.flatten()/np.sum(cm)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(names,counts,percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cm,annot = labels,cmap = colors,fmt ='')
    
    # Classification Report
    print(classification_report(y_test,classifier.predict(x_test)))
```
